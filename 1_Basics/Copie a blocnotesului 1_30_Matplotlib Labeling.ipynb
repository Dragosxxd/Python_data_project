```python
!pip install datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## 🟩 Bar Chart with Labels (1.30.1)

```python
location_counts = df['job_country'].value_counts().head(5)
plt.bar(location_counts.index, location_counts.values, color='orange')
plt.xlabel('Country')
plt.ylabel('Number of postings')
plt.title('Number of Jobs per Country')
plt.xticks(rotation = 45 ,ha = 'right')
```

---

## 🟨 Horizontal Bar Chart with Labels (1.30.2)

```python
job_avg_salary = df.groupby('job_title_short')['salary_year_avg'].mean()
job_avg_salary = job_avg_salary.sort_values(ascending = True)
plt.barh(job_avg_salary.index , job_avg_salary.values)
plt.xlabel('Average salary')
plt.title('Average salary by Job Title')
```

---

## 🟥 Line Chart with Labels (1.30.3)

```python
import datetime as dt
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['month'] = df['job_posted_date'].dt.month
avg_month = df.groupby('month')['salary_year_avg'].mean().sort_index().head(6)
avg_month
plt.plot(avg_month.index,avg_month.values,color = 'red',label = 'Average Salary')
plt.xlabel('Month')
plt.ylabel('Average Salary')
plt.title('Average Salary by Job Month')
plt.legend()

```
