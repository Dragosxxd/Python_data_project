```python
!pip install datasets
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Job Country Bar Chart (1.29.1)

```python
location_counts = df['job_country'].value_counts().head(5)
plt.bar(location_counts.index, location_counts.values, color='orange')
plt.show()
```

---

## ðŸŸ¨ Average Salary Horizontal Bar Chart (1.29.2)

```python
import matplotlib.pyplot as plt

job_avg_salary = df.groupby('job_title_short')['salary_year_avg'].mean()
job_avg_salary = job_avg_salary.sort_values(ascending = True)
plt.barh(job_avg_salary.index , job_avg_salary.values)
```

## ðŸŸ¥ Average Salary Trend Line Plot (1.29.3)

```python
import datetime as dt
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['month'] = df['job_posted_date'].dt.month
avg_month = df.groupby('month')['salary_year_avg'].mean().sort_index()
avg_month
plt.plot(avg_month.index,avg_month.values,color = 'pink')

```
