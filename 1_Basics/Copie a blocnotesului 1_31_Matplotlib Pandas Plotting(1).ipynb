```python
!pip install datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## üü© Job Posting Location Bar Plot (1.31.1)

```python
job_counts = df['job_via'].value_counts().head(5)
job_counts.plot(kind='bar')
plt.show()
```

---

## üü© Job Postings Trend Line Plot (1.31.2)

```python

df['job_posted_month'] = df['job_posted_date'].dt.to_period('M')
monthly_job_counts = df['job_posted_month'].value_counts().sort_index()
monthly_job_counts.plot(kind='line')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.title('Trend of Job Postings Over Time')
plt.show()
```

---

## üü® Top Job Locations (1.31.3)

```python
job_count = df.groupby('job_location')['job_posting'].size().sort_values(ascending = False).head(5)
job_count.plot(kind = 'barh')
plt.xlabel('Count of Jobs')
plt.ylabel('Job Location')
plt.title('Number of postings by job location')
```
